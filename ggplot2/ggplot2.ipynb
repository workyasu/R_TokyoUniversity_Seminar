```r
library(ggplot2)
```

```r
t=read.csv("HR_comma_sep.csv",header=T)
```

```r
i=0
d=t
for(i in c(3,5,6,7,8)){
    d[,c(i)]=as.factor(t[,c(i)])
}
```

```r
ggplot(data=d,aes(x=satisfaction_level))+geom_bar()
```

```r
ggplot(data=d,aes(x=satisfaction_level,fill=left))+geom_bar()
```

```r
ggplot(data=d,aes(x=satisfaction_level,fill=left))+
geom_bar()+facet_wrap(~salary,nrow=1)
```

```r
ggplot(data=d,aes(x=satisfaction_level,fill=left))+
geom_bar()+facet_grid(sales~salary)
```

```r
ggplot(data=d,aes(x=satisfaction_level,colour=left))+
geom_freqpoly()
```

```r
ggplot(data=d,aes(x=satisfaction_level,colour=sales))+
geom_freqpoly()
```
