```r
#データの読み込み
t=read.csv("HR_comma_sep.csv",header=T)
```

```r
#統計要約量
summary(t)
```

```r
i=0
d=t
for(i in c(6,7)){
    d[,c(i)]=as.factor(t[,c(i)])
}
str(d)
```

```r
# Scatter Plot
library(plotly)

set.seed(123)
ind1=sample(nrow(d),floor(nrow(d)/50))
```

```r
library(GGally)
ggpairs(d[ind1,],1:5,mapping = ggplot2::aes(color = left))
```
